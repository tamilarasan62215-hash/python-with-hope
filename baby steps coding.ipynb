BMI calculation for human 

```python
name=input("enter your name")
weight=int(input("enter your weight"))
hight=int(input("enter your hight"))
hight_meter=(hight/100)
h_m_w=(hight_meter*hight_meter)
BMI=(weight/h_m_w)
print("NAME=",name)
print("your BMI=",BMI)
if(BMI<18.5):
    print("under weight")
elif(BMI<29.9):
    print("normal")
elif(BMI<34.9):
    print("obese leval 1")
elif(BMI<39.9):
    print("obese leval 2")
else:
    print("over weight-HIGH RISK")
    
```
