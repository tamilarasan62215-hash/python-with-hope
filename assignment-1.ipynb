```python
num1=int(input("enter valu 1"))
num2=int(input("enter valu 2"))
CALC=input("add/sub/mul/div/flo div/model/power")
if(CALC=="add"):
    print("add = ",num1+num2)
elif(CALC=="sub"):
    print("sub = ",num1-num2)
elif(CALC=="mul"):
    print("mul = ",num1*num2)
elif(CALC=="div"):
    print("div = ",num1/num2)
elif(CALC=="flo div"):
    print("flo div = ",num1//num2)
elif(CALC=="model"):
    print("model = ",num1%num2)
elif(CALC=="power"):
    print("power = ",num1**num2)
else:
    print("THIS VERSION  addition, subtraction, multiplication, division, floor-division, model, and power ONLY ")
```
