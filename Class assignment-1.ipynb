```python
class OddEven():
    def OddEven():       
        z=int(input("Enter a number for even or odd"))
        if (z%2==0):
            print("Enter a number :",z)
            print(z,"is Even number")
        else:
            print(z,"is Odd number")
```

```python
class ElegibilityForMarriage():
    def Elegible():
        Gen=input("Enter your Gender")
        Age=int(input("enter your Age")) 
        if (Gen=="Male" and Age>=21):
            print("Your Gender:",Gen)
            print("Your Age :",Age)
            print("ELIGIBEL")
        elif (Gen=="Femal" and Age>=18):
            print("Your Gender:",Gen)
            print("Your Age :",Age)
            print("ELIGIBEL")
        else:
            print("Your Gender:",Gen)
            print("Your Age :",Age)
            print("NOT ELIGIBEL")
    
```

```python
Elegible1=ElegibilityForMarriage.Elegible()
```
